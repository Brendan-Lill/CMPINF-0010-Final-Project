This Notebook uses a WPRDC data set that has collected all of the fire reports since 2015. Here, each neighborhood is assessed by how many total fire reports have occured in them, with fewer being better and vice versa. 

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
df = pd.read_csv("Fire-Incidents.csv")
df.head()
neighborhood_dict = dict()
for index, row in df.iterrows():
    if pd.notna(row.iloc[9]):
        if row.iloc[9] in neighborhood_dict:
            neighborhood_dict[row.iloc[9]] += 1
        else:
            neighborhood_dict[row.iloc[9]] = 1
neighborhood_dict['Mt. Oliver'] += neighborhood_dict['Mount Oliver Borough']
del neighborhood_dict['Mount Oliver Borough']
for k, v in sorted(neighborhood_dict.items(), key=lambda x: x[1], reverse=True):
    print(k,":", v)
```

```python
data_series = pd.Series(neighborhood_dict)
sorted_series = data_series.sort_values()
bottom_10 = sorted_series.head(10)
top_10 = sorted_series.tail(10)
top_10.plot(kind='bar', color = 'red')
plt.title('Top 10 Highest Fire Neighborhoods')
plt.xlabel('Neighborhood')
plt.ylabel('Fire Reports')
plt.show()
bottom_10.plot(kind='bar', color = 'green')
plt.title('Top 10 Lowest Fire Neighborhoods')
plt.xlabel('Neighborhood')
plt.ylabel('Fire Reports')
plt.show()
```

# Conclusion

The neighborhoods that are the best according to the number of reported fires generally favor smaller, newer, and more residential areas. Out of the top 10, I have not personally visited any of them, but based on some research, they seemed like fairly nice places to live. The reason smaller areas tend to be favored is that there are fewer places where a fire could be started. Along with this, neighborhoods with newer buildings would be more resistant to fire due to new construction technology, and this list favors residential areas as they tend to do less potentially fire-causing activities compared to commercial places like restaurants and factories.
